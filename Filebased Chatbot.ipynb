```python
import os
import gradio as gr
import ollama
import pdfplumber
import pandas as pd
import pytesseract
from PIL import Image
from docx import Document

 
os.system("ollama pull llama3.2:latest")


def extract_text_from_pdf(pdf_file):
    try:
        with pdfplumber.open(pdf_file) as pdf:
            text = ''
            for page in pdf.pages:
                text += page.extract_text()
        return text
    except Exception as e:
        return f"Error reading PDF: {str(e)}"


def extract_text_from_docx(docx_file):
    try:
        doc = Document(docx_file)
        text = ''
        for para in doc.paragraphs:
            text += para.text
        return text
    except Exception as e:
        return f"Error reading DOCX: {str(e)}"


def extract_text_from_excel(excel_file):
    try:
        df = pd.read_excel(excel_file)
        return df.to_string()
    except Exception as e:
        return f"Error reading Excel: {str(e)}"


def extract_text_from_image(image_file):
    try:
        image = Image.open(image_file)
        text = pytesseract.image_to_string(image)
        return text
    except Exception as e:
        return f"Error reading image: {str(e)}"


def chat_with_llama_3_2(input_text, uploaded_file):
    try:
        
        file_text = ""
        if uploaded_file is not None:
            file_extension = uploaded_file.name.split('.')[-1].lower()
            if file_extension == 'pdf':
                file_text = extract_text_from_pdf(uploaded_file.name)
            elif file_extension == 'docx':
                file_text = extract_text_from_docx(uploaded_file.name)
            elif file_extension == 'xls' or file_extension == 'xlsx':
                file_text = extract_text_from_excel(uploaded_file.name)
            elif file_extension in ['jpg', 'jpeg', 'png']:
                file_text = extract_text_from_image(uploaded_file.name)
            else:
                file_text = "Unsupported file format."
        
        
        full_input = f"File content: {file_text}\n\nQuestion: {input_text}"
        
        
        response = ollama.chat(model="llama3.2:latest", messages=[{"role": "user", "content": full_input}])
        
        
        assistant_message = response.get('message', {}).get('content', 'Error: No response content found')
        
        return assistant_message
    
    except Exception as e:
        return f"Error: {str(e)}"


iface = gr.Interface(
    fn=chat_with_llama_3_2,
    inputs=[gr.Textbox(label="Ask a Question"), gr.File(label="Upload a File", type="filepath")],
    outputs="text",
    title="LLaMA 3.2 File-based Q&A Chatbot (By: Ronit Roy)"
)


iface.launch()
```
